#### Dictionaries are unordered collections of items.
#### They store data in key-value pairs. Keys must be unique and immutable, while values can be of any type.

```python
student = {'name': 'Uday', 'age': 20, 'marks': 29}

print(student)
print(student['marks'])
print(student.get('age'))
```

```python
student['age'] = 21 # updating value
print(student)

student['country'] = 'India' # adding a new key and value
print(student)

del student['marks']
print(student)
```

```python
# Dictionary Methods

keys = student.keys()
print(keys)

values = student.values()
print(values)

key_value = student.items()
print(key_value)
```
