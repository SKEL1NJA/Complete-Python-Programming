## Text Files

```python
with open("example.txt", "r") as file:
    content = file.read()
    print(content)
```

```python
with open("example.txt", "r") as file:
    for line in file:
        print(line.strip()) # strip remove newline character
```

```python
with open("example.txt", "w") as file:
    file.write("This will remove all the data and this line")
```

```python
with open("example.txt","a") as file:
    file.write("\nThis will append the data")
```

```python
lines=['First line \n','Second line \n','Third line\n']

with open("example.txt","w") as file:
    file.writelines(lines)
```
