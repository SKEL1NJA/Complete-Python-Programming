#### Lists are ordered, mutable collections of items.
#### They can contain items of different data types.
#### use []

```python
lst = [13, 'Uday', 3.14, False]

print(lst[1])
print(lst[-1])

print(lst[0:2])
print(lst[0:4:2])
print(lst[::-1])

lst[0] = 7
print(lst)
```

```python
fruits = ['apple', 'banana', 'banana', 'cherry', 'guava']
print(fruits)

# List Methods
fruits.append('Kiwi')
print(fruits)

fruits.insert(1, 'watermelon')
print(fruits)

fruits.remove('guava')
print(fruits)

popped_fruit = fruits.pop()
print(popped_fruit)
print(fruits)

idx = fruits.index('cherry')
print(idx)

print(fruits.count('banana'))

fruits.sort()
print(fruits)

fruits.reverse()
print(fruits)

fruits.clear()
print(fruits)
```

```python
numbers = [2, 4, 6, 8, 10]

for number in numbers:
    print(number)
```

```python
for index, number in enumerate(numbers):
    print(index, number)
```

### List Comprehension

```python
# Basic list comprehension
square = [num**2 for num in range(10)]
print(square)
```

```python
# with conditional logic
even_square = [num**2 for num in range(10) if num%2 == 0]
print(even_square)
```

```python
# Nested List Comprehension
lst1 = [1,2,3,4]
lst2 = ['a','b','c','d']

pair = [[i, j] for i in lst1 for j in lst2]
print(pair)
```

```python
## List Comprehension with function calls
words = ["hello", "world", "python", "list", "comprehension"]
length = [len(word) for word in words]
print(length)
```
