### a+b -> np.add(a, b)
### a-b -> np.subtract(a, b)
### a*b -> np.multipy(a, b)
### a/b -> np.divdide(a, b)
### a%b -> np.mod(a, b)
### a**b -> np.power(a, b)
### 1/a -> np.reciprocal(a, b)

#### can perform these operations manually as well

```python
import numpy as np
```

```python
a = [[10, 2, 8, 15],
     [3, 12, 1, 5]]

b = [[1,2,3,4],
     [1,2,3,4]]
```

```python
print(np.add(a,b))
```

```python
print(np.subtract(a,b))
```

```python
print(np.multiply(a,b))
```

```python
print(np.divide(a,b))
```

```python
print(np.mod(a,b))
```

```python
print(np.power(a,b))
```

```python
print(np.reciprocal(a))
```
