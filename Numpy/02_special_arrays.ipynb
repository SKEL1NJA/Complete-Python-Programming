```python
import numpy as np
```

```python
# array filled with all 0's

arr0 = np.zeros(4)
print(arr0)

print()

arr0_1 = np.zeros((4,4))
print(arr0_1)
```

```python
# array filled with all 1's

arr1 = np.ones((3,3))
print(arr1)
```

```python
# empty array -- memory may already contain old data from previous programs.

arr_empty = np.empty(4)
print(arr_empty)
```

```python
# array within a range

arr_range = np.arange(4)
print(arr_range)
```

```python
# identity matrix

arr_identity = np.eye(3)
print(arr_identity)
```

```python
# array in which values are linearly spaced in a specif interval

arr_lin = np.linspace(0, 10, 3)
print(arr_lin)
```
