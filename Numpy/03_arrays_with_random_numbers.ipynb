```python
import numpy as np
```

### rand(): random values between 0 & 1

```python
np.random.rand(4,3)
```

### randn() --- close to 0 values, maybe +ve, -ve as well

```python
np.random.randn(4,3)
```

### ranf() --- [0.0, 1.0)

```python
np.random.ranf((4,3))
```

### randint(start, end, how_many)

```python
np.random.randint(4,50,7)
```
