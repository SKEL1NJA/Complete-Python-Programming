### np.min(x)
### np.max(x)
### np.argmin(x)/np.argmax(x) -- returns the index of minimum/maximum element
### np.sqrt(x)
### np.sin(x)
### np.cos(x)
### np.cumsum(x) -- works like prefix sum logic of cpp

### in 2d array -- axis = 0 --> along column, axis = 1 --> along row

```python
import numpy as np
```

```python
x = [1, 2, 3, 4, 5]
print(np.min(x), np.argmin(x))
print(np.max(x), np.argmax(x))
print(np.sqrt(x))
print(np.sin(x))
print(np.cos(x))
print(np.cumsum(x))
```

```python
x = [[1,6,3,8],
     [5,2,7,4]]

print(np.min(x, axis=1))
print(np.min(x, axis=0))
print()
print(np.max(x, axis=1))
print(np.max(x, axis=0))
```
